##  Visual Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
#load data using books_and_reviews.csv

books_df = pd.read_csv('books_and_reviews.csv')
books_df.head(20)
```

```python
#load data using Google data

booksg_df = pd.read_csv('new_google_api_dataset.csv')
booksg_df.head(20)
```

```python
#Histogram: Plot the distribution of ratings to see the frequency of each rating. 
#This can help understand if the reviews are generally positive, negative, or mixed.

plt.figure(figsize=(10, 6))
sns.histplot(books_df['review_score'], bins=10, kde=False, color='skyblue')
plt.title('Distribution of Ratings (Books DF)')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(booksg_df['average_rating'], bins=10, kde=False, color='skyblue')
plt.title('Distribution of Ratings (Google Data)')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
#scatter plot to see if there is a correlation between how long a book is and the rating - doesn't appear to be any correlation

booksg_df.plot.scatter(x='average_rating', y='page_count', title='Relationship between Page Count and Ratings')
```

```python
#top publishers
google_books_df = pd.read_csv('new_google_api_dataset.csv')
top_10_publishers = google_books_df.groupby('publisher')['title'].count().sort_values(ascending=False).head(10)

#top_10_publishers 
top_10_publishers.plot(kind='bar')
```

```python

# Load the dataset
booksg_df = pd.read_csv('new_google_api_dataset.csv')

# Calculate the top 10 publishers based on the number of titles
#top_10_publishers = booksg_df['publisher'].value_counts().head(10)

top_10_publishers = booksg_df.groupby('publisher')['title'].count().sort_values(ascending=False).head(10)

# Filter the dataset to include only books from the top 10 publishers
top_publishers_df = booksg_df[booksg_df['publisher'].isin(top_10_publishers.index)]

# Calculate the average rating for each of the top 10 publishers
avg_ratings = top_publishers_df.groupby('publisher')['average_rating'].mean()

# Plot the average ratings of the top 10 publishers
plt.figure(figsize=(10, 6))
avg_ratings.plot(kind='bar', color='skyblue')
plt.xlabel('Publishers')
plt.ylabel('Average Rating')
plt.title('Average Rating by Top 10 Publishers')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Calculate the top 10 categories based on the number of titles
top_10_categories = books_df['categories'].value_counts().head(10)

# Filter the dataset to include only books from the top 10 categories
top_categories_df = books_df[books_df['categories'].isin(top_10_categories.index)]

# Calculate the average rating for each of the top 10 categories
avg_ratings = top_categories_df.groupby('categories')['review_score'].mean()

# Plot the average ratings of the top 10 publishers
plt.figure(figsize=(10, 6))
avg_ratings.plot(kind='bar', color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Average Rating')
plt.title('Average Rating by Top 10 Categories')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Comparative Analysis - Box plots can show the median, quartiles, and outliers in the ratings.

top_10_categories = books_df['categories'].value_counts().head(10)

# Filter the dataset to include only books from the top categories (genres)
top_categories_df = books_df[books_df['categories'].isin(top_10_categories.index)]

# Calculate the average rating for each of the top categories (genres)
avg_ratings = top_categories_df.groupby('categories')['review_score'].mean()
plt.figure(figsize=(12, 6))
sns.boxplot(x='review_score', y='categories', data=top_categories_df)
plt.title('Comparison of Ratings Across Books')
plt.show()

"""

For each category, there is a boxplot that represents the distribution of review scores:

The box shows the interquartile range (IQR) of the data, which contains the middle 50% of the review scores for that category.
The line represents the median review score for that category.
The "whiskers" extending from the box show the range of the data, excluding outliers.
Outliers: Any points outside of the whiskers are considered outliers and are plotted individually.

The 'Computers' category has a very narrow range of review scores, 
suggesting that the reviews in this category are very consistent or that maybe reviewers are generally consistent with their ratings.
"""
```

```python
#Bar chart to visialise the number of reviews for each category to see which genres are most reviewed.

plt.figure(figsize=(12,8))
category_counts = top_categories_df['categories'].value_counts()
sns.barplot(x=category_counts.values, y=category_counts.index)
plt.title('Number of Reviews by Category')
plt.xlabel('Number of Reviews')
plt.ylabel('Categories')
plt.show()
```

```python
#Top 50 categories and top rated books - split to top 50 as the data was very big

top_50_categories = books_df['categories'].value_counts().head(50)

# Filter the dataset to include only books from the top categories (genres)
top_50_categories_df = books_df[books_df['categories'].isin(top_50_categories.index)]


top_rated_books = books_df.loc[top_50_categories_df.groupby('categories')['review_score'].idxmax()]
plt.figure(figsize=(14,8))
sns.barplot(x='review_score', y='categories', data=top_rated_books, hue='title', dodge=False)
plt.title('Top-Rated Book in Each Category')
plt.xlabel('Review Score')
plt.ylabel('Categories')
plt.show()
```

```python
#Word Cloud: Create word clouds to visualiise the most common words used in reviews for each category.

from wordcloud import WordCloud

categories = top_categories_df['categories'].unique()
for category in categories:
    text = " ".join(review for review in books_df[books_df['categories'] == category]['review_text'])
    wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
    
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"Word Cloud for {category}")
    plt.show()
```
