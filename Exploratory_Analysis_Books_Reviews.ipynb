## Exploratory Analysis of Book and Reviews Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 
import seaborn as sns
```

```python
df = pd.read_csv('books_and_reviews.csv')
df.head(20)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False,yticklabels=False)
plt.title('Missing Data Heatmap')
plt.show()
```

```python
#Review Length Analysis: Distribution of length of reviews

plt.figure(figsize=(10, 6))
sns.histplot(df['review_score'], bins=10, kde=True, color='blue')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
df.review_score.value_counts()
```

```python
# Show top 20 authors by mean rating
mean_rating_by_author = df.groupby('authors')['review_score'].mean().sort_values(ascending=False)
print(mean_rating_by_author.head(20))
```

```python
#Identify Outliers in Mean Ratings
#Identify books or authors whose mean ratings are outliers compared to the rest of the dataset.

highly_rated_books = df.groupby('title')['review_score'].mean().sort_values(ascending=False).head(10)
print("Top 10 Highly Rated Books:")
print(highly_rated_books)
```
